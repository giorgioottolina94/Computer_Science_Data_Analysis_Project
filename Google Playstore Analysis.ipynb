# Convert App Sizes to a number and Transform “Varies with device” into a missing value

```python
df['Size'].value_counts()
```

```python
df['Size'] = df.Size.replace("Varies with device",np.nan)
df['Size'] = df.Size.str.replace("M","000") # All size values became the kilobyte type.
df['Size'] = df.Size.str.replace("k","")
df['Size'] = df.Size.replace("1,000+",1000)
df['Size'] = df['Size'].astype(float)
df['Size'].dtype
```

# Create two dataframes: one for the genres and one bridging apps and genders. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
Apps = []
GenreList = []

for index, row in df.iterrows():
    GenresApp = row['Genres'].split(';')
    for genre in GenresApp:
        GenreList.append(genre)
        Apps.append(row['App'])
```

```python
GenresDF = pd.DataFrame({'Genres' : list(set(GenreList))})
```

```python
GenresDF
```

```python
app_genre = pd.DataFrame({'App' : Apps, 'GenresSeparated' : GenreList})
```

```python
app_genre
```

# For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
falseVector = [False] * df.shape[0]
GenreList = list(set(GenreList))

for genre in GenreList:
    df[genre] = falseVector
    
for index, row in df.iterrows():
    differentGenres = row['Genres'].split(';')
    for genre in differentGenres:
        df.loc[index, genre] = True
```

```python
df
```
